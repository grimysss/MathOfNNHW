```python
import numpy as np 
from matplotlib import pylab
import pylab as plt
```

```python
def sigmoid(x):
    s = 1/(1+np.exp(-x))
    return s

print(sigmoid(0.45))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s*(1-s)

print(sigmoid_derivative(1))
```

```python
values = plt.linspace(-10,10,100)
plt.rcParams['figure.figsize'] = [10, 5]
plt.title("sigmoid and sigmoid derivative")
plt.plot(values, sigmoid(values), "b", label="sigmoid")
plt.plot(values, sigmoid_derivative(values), "--r", label="sigmoid derivative")
plt.legend(loc="best")
plt.grid()
```

```python
#Tanh
def tanh(x):
    return np.tanh(x)

def tanh_derivative(x):
    return (1 - (np.tanh(x)** 2))
```

```python
values = plt.linspace(-10,10,100)
plt.rcParams['figure.figsize'] = [10, 5]
plt.title("hyperbolic tangent and hyperbolic tangent derivative")
plt.plot(values, tanh(values), "b", label="tanh")
plt.plot(values, tanh_derivative(values), "--r", label="tanh derivative")
plt.legend(loc="best")
plt.grid()
```
