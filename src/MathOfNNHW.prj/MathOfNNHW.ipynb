```python
# Финкция сигмойды.


def sigmoid(x):
    s = 1/(1+np.exp(-x))
    return s


# Производная функция сигмойды.

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s*(1-s)
```

```python
# Строим сигмойду и производную функции сигмойды.

values = plt.linspace(-10, 10, 100)
plt.rcParams['figure.figsize'] = [15, 10]
plt.title("sigmoid and sigmoid derivative")
plt.plot(values, sigmoid(values), "b", label="sigmoid")
plt.plot(values, sigmoid_derivative(values), "--r", label="sigmoid derivative")
plt.legend(loc="best")
plt.grid()
```

```python
# Функция гиперболоидного тангенса.


def tanh(x):
    return np.tanh(x)

# Производная функция гиперболоидного тангенса.


def tanh_derivative(x):
    return (1 - (np.tanh(x) ** 2))
```

```python
# Строим гиперболоидного тангенс и производная функция гиперболоидного тангенса.

values = plt.linspace(-10, 10, 100)
plt.rcParams['figure.figsize'] = [15, 10]
plt.title("hyperbolic tangent and hyperbolic tangent derivative")
plt.plot(values, tanh(values), "b", label="tanh")
plt.plot(values, tanh_derivative(values), "--r", label="tanh derivative")
plt.legend(loc="best")
plt.grid()
```
